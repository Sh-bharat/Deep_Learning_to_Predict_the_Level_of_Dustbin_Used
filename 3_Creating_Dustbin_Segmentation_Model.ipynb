```python
model=YOLO('/content/Dustbin_Segmentation_Weights/segment/train/weights/last.pt')
```

```python
image_folder = '/content/Test'
image_files = [f for f in os.listdir(image_folder) if f.endswith(('.jpg', '.jpeg', '.png'))]

num_rows = -(-len(image_files) // 5)
fig, axs = plt.subplots(num_rows, 5, figsize=(20, 4 * num_rows))
axs = axs.flatten()

for i, image_file in enumerate(image_files):
    image_path = os.path.join(image_folder, image_file)
    results = model(image_path)
    segmented_image = Image.fromarray(results[0].plot())
    axs[i].imshow(segmented_image)
    axs[i].axis('off')

plt.tight_layout()
plt.show()
```

```python
input_dir = '/content/drive/MyDrive/Waste_segmentation/Vedios'
output_dir = '/content/drive/MyDrive/Waste_segmentation/Output_Vedios'
count=1

for file in os.listdir(input_dir):
    if file.endswith('.mp4'):
        file_path = os.path.join(input_dir, file)
        output_file_path=os.path.join(output_dir,"Output_"+file)
        output_frames=[]
        if (os.path.exists(file_path)):
            vedio = cv2.VideoCapture(file_path)
            print(f"Taken Vedio Count {count}")
            count=count+1

            success,image=vedio.read()
            height, width, layers = image.shape
            size = (width,height)
            fps = vedio.get(cv2.CAP_PROP_FPS)
            frame_interval = int(fps)

            while success:
                success, image = vedio.read()
                if success:
                    results = model(image)
                    segmented_frame = results[0].plot()
                    # Convert the PIL Image to a NumPy array before appending to output_frames
                    output_frames.append(np.array(Image.fromarray(segmented_frame)))

            out = cv2.VideoWriter('Output'+str(count-1)+'.avi',cv2.VideoWriter_fourcc(*'DIVX'), fps, size)
            for i in range(len(output_frames)):
                out.write(output_frames[i])
            out.release()
            print(f'Output Vedio Saved.')
            if(os.path.exists(output_dir)):
              print(f'Output Directory Found')
            else:
              os.mkdir(output_dir)
              print(f'Output Directory Created')
            shutil.copyfile('/content/'+'Output'+str(count-1)+'.avi',output_dir+'/Output'+str(count-1)+'.avi')

        else:
            raise IOError("Error: File  not Found. ")
```
